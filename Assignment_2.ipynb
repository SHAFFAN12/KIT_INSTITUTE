# Cleaning Data

```python
Movies_data["MOVIES"] = Movies_data['MOVIES'].str.strip()
```

# Cleaning the votes

```python
Movies_data.isnull().sum()
```

```python
Movies_data['VOTES']=Movies_data['VOTES'].fillna('0')
```

```python
Movies_data['VOTES']=Movies_data['VOTES'].str.replace(',','')
```

```python
Movies_data['VOTES']=Movies_data['VOTES'].astype(int)
```

```python
Movies_data.info()
```

# Rating

```python
Movies_data['RATING'] = Movies_data['RATING'].fillna(0.0)
```

```python
Movies_data.isnull().sum()
```

# **Cleaning** **Genre** **bold**

```python
Movies_data['GENRE'] = Movies_data['GENRE'].str.strip('\n')
```

```python
Movies_data[['Type 1','Type 2','Type 3']] = Movies_data['GENRE'].str.split(',',2,expand=True)
```

# Cleaning the yera column

```python
Movies_data['YEAR'] = Movies_data['YEAR'].str.replace('(','')
```

```python
Movies_data['YEAR']
```

```python
Movies_data[['YEAR','Random']]=Movies_data['YEAR'].str.split('-',2,expand=True)
```

```python
Movies_data
```

```python
Movies_data['Gross in million $']=Movies_data['Gross in million $'].str.replace('$','').str.replace('M','')
```

```python
Movies_data['Gross in million $'].dtype
```

```python
Movies_data['Gross in million $'] = Movies_data['Gross in million $'].astype(float)
```

```python
Movies_data['Gross in million $'].dtype
```

```python
Movies_data['Gross in million $']=Movies_data['Gross in million $'].fillna(0.0)
```

# **Stars columns**
