Creating label encoding and bins


```python
import pandas as pd
```

```python
data = pd.DataFrame({
    "Name":["Alice","Bob","Charlie","John","Alice"],
    "Salary":[45000,55000,61000,71000,100000]
})
```

```python
data
```

```python
data['Salary Category'] = pd.cut(data['Salary'], bins=[0,50000,60000,70000,90000,float("inf")], labels=["Very low","low","Medium","High","Very High"])
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
enc = LabelEncoder()
data['Name'] = enc.fit_transform(data['Name'])
data['Salary Category'] = enc.fit_transform(data['Salary Category'])
```

```python
data
```
